# BEDP vs Uniform on 8×8 Heisenberg Model

This notebook compares **BEDP** (BrQin Entropy-Direction Paradigm with CE/LP switching) vs **uniform growth** on a 8×8 Heisenberg model.

Goal: Show faster convergence, lower energy, and richer structure with BEDP.

```python
import numpy as np
import matplotlib.pyplot as plt
from oracle_peps_sci import PepsOracleSci
```

```python
# Run uniform baseline
oracle_uniform = PepsOracleSci(Lx=8, Ly=8, steps=50)
metrics_uniform = oracle_uniform.run(guided=False)

# Run BEDP guided
oracle_guided = PepsOracleSci(Lx=8, Ly=8, steps=50)
metrics_guided = oracle_guided.run(guided=True)

print("Uniform:", metrics_uniform)
print("BEDP:", metrics_guided)
```

```python
# Plot comparison
plt.figure(figsize=(10, 6))
plt.plot(range(1, 51), [oracle_uniform.compute_heisenberg_energy() for _ in range(50)], label='Uniform', alpha=0.7)
plt.plot(range(1, 51), [oracle_guided.compute_heisenberg_energy() for _ in range(50)], label='BEDP (CE/LP)', alpha=0.7)
plt.xlabel('Step')
plt.ylabel('Energy')
plt.title('BEDP vs Uniform on 8×8 Heisenberg')
plt.legend()
plt.grid(True)
plt.show()
```
